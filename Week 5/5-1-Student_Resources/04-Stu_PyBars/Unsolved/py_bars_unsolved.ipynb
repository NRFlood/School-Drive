```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
```

```python
# Create a bar chart based upon the above data
plt.bar(x_axis, bars_in_cities)
plt.show
```

```python
# Create the ticks for our bar chart's x axis
plt.bar(x_axis, bars_in_cities)
plt.xticks(x_axis, cities, rotation=90)
plt.show
```

```python
# Set the limits of the x axis
plt.bar(x_axis, bars_in_cities)
plt.xticks(x_axis, cities, rotation=90)
plt.xlim(-0.75, len(x_axis)-0.25)
plt.show
```

```python
# Set the limits of the y axis
plt.bar(x_axis, bars_in_cities)
plt.xticks(x_axis, cities, rotation=90)
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(bars_in_cities)*1.1)
plt.show
```

```python
# Give the chart a title, x label, and y label
plt.bar(x_axis, bars_in_cities)
plt.xticks(x_axis, cities)
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, 9)
plt.xlabel('Cities')
plt.ylabel('Number of Bars Per 100 people')
plt.title('Bars Per Capita (100K)')
plt.show
```

```python
# Save an image of the chart and print it to the screen
plt.bar(x_axis, bars_in_cities)
plt.xticks(x_axis, cities)
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, 9)
plt.xlabel('Cities')
plt.ylabel('Bars')
plt.title('Bars Per Capita (100K)')
plt.savefig('../Images/bars.png')
plt.show()
```
